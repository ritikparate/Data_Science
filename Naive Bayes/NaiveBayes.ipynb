# Dataset is now ready for Model Building

# Model Building

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(xtrain,ytrain)
```

```python
ytrain_pred=model.predict(xtrain)
```

```python
ytest_pred=model.predict(xtest)
```

# Evaluating

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytrain,ytrain_pred) #### Accuracy on train data
```

```python
accuracy_score(ytest,ytest_pred)   #### Accuracy on test data
```

We got Training accuracy of 80.12 and Testing accuracy of 79.85, there is very small difference between training accuracy and testing accuracy.

Therefore we can conclude that the model is not overfitted.

# Check Null Accuracy

```python
ytest.value_counts()
```

```python
null_accuracy = (11360/(11360+3700))

print('Null accuracy score: {0:0.4f}'. format(null_accuracy))
```

We can see that our model accuracy score is 80% and null accuracy score is 75.5%.

So, we can conclude that our Naive Bayes Classification model is doing a very good job in predicting the class labels.

# Confusion matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(ytest, ytest_pred)
```

```python
# visualize confusion matrix with seaborn heatmap

cm_matrix = pd.DataFrame(data=cm, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='.5g')
plt.show()
```
